```python
# all imports will go here 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('./train_dataset.csv')
salaries = pd.read_csv('./train_salaries.csv')
print(f"Shape of training and salaries are {train.shape}, {salaries.shape} respectively")
```

```python
train.head()
```

```python
df = pd.merge(train, salaries, on="jobId")
```

```python
df.head()
```

```python
# see the unique values present in df 
df.nunique()
```

# let's explore the dataset 


```python
# lets see the distribution of the salaries 
plt.subplots(1, 1, figsize = (16, 5))
sns.kdeplot(df['salary'], shade=True, palette="vlag")
plt.title("Salary (PAISA)", size=20)
plt.show()
```

**Salary looks so sweetly normalised**
 

```python
# let's see the distributoin for the yearsofexperience variable 
plt.subplots(1, 1, figsize = (16, 5))
sns.kdeplot(df['yearsExperience'], shade=True, palette="vlag")
plt.title("yearsExperience (MEHANAT)", size=20)
plt.show()
```

```python
plt.subplots(1, 1, figsize = (16, 5))

sns.countplot(df['jobType'], palette="vlag")
plt.show()
```

```python
plt.subplots(1, 1, figsize = (16, 5))

sns.countplot(df['degree'], palette="vlag")
plt.show()
```

```python
plt.subplots(1, 1, figsize = (16, 5))

sns.countplot(df['major'], palette="vlag")
plt.show()
```

```python
#industry

plt.subplots(1, 1, figsize = (16, 5))

sns.countplot(df['industry'], palette="vlag")
plt.show()
```
